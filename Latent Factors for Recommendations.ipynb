```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
import random
from numpy.linalg import norm
```

```python
iterations = 40
k = 20
regularization_factor = 0.1 # λ
learning_rate = 0.01 #n
data_path = "ratings.csv"
```

```python
# stocahstic gradient descent function for Latent factors for recommendations
def latentFactors(data_path, learning_rate, regularization_factor, k, iterations):
    #coef = [0.0 for i in range(len(train[0]))]
    # dictionary for movie matrix
    Q = {}
    # dictionary for user matrix
    P = {}
    # initialize array of iteration errors with 0's
    iteration_errors = [0.0] * iterations
    for iteration in range(iterations):
        # Open file        
        fileHandler = open (data_path, "r")
        # each iteration
        while True: 
            # Get next line from file
            line = fileHandler.readline()
            # If line is empty then end of file reached
            if not line :
                break;
                
            # trim each row
            row = line.strip().split(',')
            movie_id = row[0]
            user_id = row[1]
            rating = int(row[2])
            
            # initialize each element in row of Q random value in [0, (5/k)^1/2]
            if movie_id not in Q:
                Q[movie_id] = [0.0] * k
                Q[movie_id] = [x + random.uniform(0, sqrt(5/k)) for x in Q[movie_id]]

            # initialize each element in row of Q random value in [0, (5/k)^1/2]    
            if user_id not in P:
                P[user_id]  = [0.0] * k
                P[user_id] = [x + random.uniform(0, sqrt(5/k)) for x in P[user_id]]
    
            # values from previous iteration
            old_movie = Q[movie_id]
            old_user = P[user_id]
            
            # derivative of error
            error = 2 * (rating - np.dot(old_movie, np.transpose(old_user)))

            # update each element of row of Q
            for k_fraction in range(len(Q[movie_id])):
                Q[movie_id][k_fraction] += learning_rate * (error*old_user[k_fraction]
                                                         - (regularization_factor * old_movie[k_fraction]))
            # update each element of row of P
            for k_fraction in range(len(P[user_id])):
                P[user_id][k_fraction] += learning_rate * (error*old_movie[k_fraction]
                                                         - (regularization_factor * old_user[k_fraction]))
        # Close file
        fileHandler.close()
        
        # Open file        
        fileHandler = open (data_path, "r")
        # initialize variables to store each part of error equation
        # use these variables to calculate error of each iteration at the end of iteration
        # error: ∑ (Riu - Qi * Transpose(Pu))^2
        total_error = 0.0
        #regularization of Pu: ∑ pow(L2_norm(Pu)) 
        total_reg_P = 0.0
        #regularization of Qi: ∑ pow(L2_norm(Qi)) 
        total_reg_Q = 0.0
        
        while True: 
            # Get next line from file
            line = fileHandler.readline()
            # If line is empty then end of file reached
            if not line :
                break;

            # trim each row
            row = line.strip().split(',')
            movie_id = row[0]
            user_id = row[1]
            rating = int(row[2])
            
            # sum each part of error equation
            # ∑ (Riu - Qi * Transpose(Pu))^2
            hypothesis_rating = np.dot(Q[movie_id], np.transpose(P[user_id]))
            total_error += pow((rating - hypothesis_rating), 2)
            
        # Close file
        fileHandler.close()
        
        #regularization of Pu: ∑ pow(L2_norm(Pu)) 
        for user in P:
            total_reg_P +=  pow(norm(P[user]), 2)
        
        #regularization of Qi: ∑ pow(L2_norm(Qi)) 
        for movie in Q:
            total_reg_Q = pow(norm(Q[movie]), 2)

        iteration_errors[iteration] = total_error + (regularization_factor *(total_reg_P + total_reg_Q))
        
    return iteration_errors
```

```python
learning_rate = 0.001
result = latentFactors(data_path, learning_rate, regularization_factor, k, iterations)
result
```
